```python
df = pd.read_excel('Data/07-05-2024_TextNorm.xlsx')
df
```

```python
bd = [str(i) for i in df['Birthday'].values]

years = []
for d in bd:
  d = d.replace('/', '*').replace('-', '*')
  d_mod = d.split('*')
  for s in d_mod:
    if len(s) == 4:
      years.append(s)
      break

age = []
for year in years:
  age.append(2024-int(year))

df['Age'] = np.array(age)
```
